```python
def parse_int(s, base=10):
    return int(s, base)
```

```python
parse_int("24", 8)
```

```python
#ValueError
parse_int("234.23")
```

```python
#TypeError
a = None
parse_int(a)
```

```python
def try_parse_int(s, base=10, val=0): 
    try:
        a = int(s, base)
        return a
    except (ValueError, TypeError):
        print("Type/Value Error")
        return val
```

```python
try_parse_int('24', 8)
```

```python
s = "25.56"
try_parse_int(s)
```

```python
s = None
try_parse_int(s)
```

```python
iris = load_iris()
X = iris.data
y = iris.target

# Pick 3 features for 3D plot
x1 = X[:, 0]  # sepal length
x2 = X[:, 1]  # sepal width
x3 = X[:, 2]  # petal length

# Plot 3D Scatterplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x1, x2, x3, c=y, cmap='viridis', s=50)

ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
plt.title("3D Scatter Plot of Iris Data")
plt.legend(*scatter.legend_elements(), title="Species")
plt.show()
```

```python
# 3D wireframe
# Create grid
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # A surface function

# Plot wireframe
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='gray')
ax.set_title("3D Wireframe Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
# Plot surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title("3D Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()
```

```python
from scipy import stats
fig = plt.figure()
ax1 = fig.add_subplot(211)
np.random.seed(0)
x = stats.loggamma.rvs(5, size=500) + 5
# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution
# ANSWER

prob = stats.probplot(x, dist=stats.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
ax2 = fig.add_subplot(212)
xt, lamda = stats.boxcox(x)
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
```

```python
import geopandas as gpd
import requests
import zipfile
import os

# Download countries data from here and save in the same folder as notebook
# url = "https://www.naturalearthdata.com/downloads/110m-cultural-vectors/110m-admin-0-countries/"
zip_path = "ne_110m_admin_0_countries.zip"
shp_dir = "ne_110m_admin_0_countries"

if not os.path.exists(shp_dir):
    
    # Extract zip
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(shp_dir)
```

```python
# Load shapefile
world = gpd.read_file(os.path.join(shp_dir, 'ne_110m_admin_0_countries.shp'))

# Filter only Australia
australia = world[world['ADMIN'] == 'Australia']

# Australian cities with coordinates
data = {
    'City': ['Sydney', 'Melbourne', 'Brisbane', 'Perth', 'Adelaide'],
    'Latitude': [-33.8688, -37.8136, -27.4698, -31.9505, -34.9285],
    'Longitude': [151.2093, 144.9631, 153.0251, 115.8605, 138.6007]
}

cities_df = pd.DataFrame(data)

# Convert cities to GeoDataFrame
cities_gdf = gpd.GeoDataFrame(
    cities_df, 
    geometry=gpd.points_from_xy(cities_df.Longitude, cities_df.Latitude),
    crs="EPSG:4326"
)

# Plot Australia map and cities
fig, ax = plt.subplots(figsize=(10, 8))
australia.plot(ax=ax, color='lightgray', edgecolor='black')
cities_gdf.plot(ax=ax, color='red', markersize=50)

# Label cities
for x, y, label in zip(cities_gdf.geometry.x, cities_gdf.geometry.y, cities_gdf['City']):
    ax.text(x + 0.3, y + 0.3, label, fontsize=9)

plt.title("Major Australian Cities")
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind, f_oneway
import matplotlib.pyplot as plt

np.random.seed(0)

# Simulate test scores for 3 groups (same mean)
group_A = np.random.normal(loc=70, scale=10, size=30)
group_B = np.random.normal(loc=70, scale=10, size=30)
group_C = np.random.normal(loc=70, scale=10, size=30)

# Combine into a DataFrame
df = pd.DataFrame({
    'score': np.concatenate([group_A, group_B, group_C]),
    'group': ['A']*30 + ['B']*30 + ['C']*30
})
df
```

```python
# === Step 1: Multiple t-tests ===
t_AB = ttest_ind(group_A, group_B)
t_BC = ttest_ind(group_B, group_C)
t_AC = ttest_ind(group_A, group_C)

print("\n🔹 Multiple t-tests (each p-value):")
print(f"A vs B: p = {t_AB.pvalue:.4f}")
print(f"B vs C: p = {t_BC.pvalue:.4f}")
print(f"A vs C: p = {t_AC.pvalue:.4f}")
```

```python
# === Step 2: One-way ANOVA ===
anova_result = f_oneway(group_A, group_B, group_C)
print(f"\n🔹 One-way ANOVA: F = {anova_result.statistic:.4f}, p = {anova_result.pvalue:.4f}")

# === Optional: Boxplot ===
plt.figure(figsize=(6, 4))
df.boxplot(by='group', column='score', grid=False)
plt.title("Boxplot: Scores by Teaching Method")
plt.suptitle('')
plt.xlabel("Teaching Method")
plt.ylabel("Test Score")
plt.show()
```
