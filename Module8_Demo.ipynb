```python
# Regular Expressions
import re

# Sample text
text = "The rain in Spain falls mainly in the plain. Spain is sunny."
```

```python
# 1️⃣ split() - split text by spaces or punctuation
print("split():", re.split(r"\s+", text))  # split by whitespace
```

```python
# 2️⃣ search() - find first occurrence of 'Spain'
match_search = re.search(r"Spain", text)
if match_search:
    print("search(): Found 'Spain' at position", match_search.start())
```

```python
# 3️⃣ findall() - find all words ending with 'ain'
print("findall():", re.findall(r"\w*ain\b", text))

# 4️⃣ finditer() - same as findall but returns iterator of match objects
print("finditer():")
for match in re.finditer(r"\w*ain\b", text):
    print(" -", match.group(), "at", match.start())
```

```python
# 5️⃣ match() - match at the start of the string
match_match = re.match(r"Spain", text)
if match_match:
    print("match():", match_match.group(), "found at start")
```

```python
# Using groups() example
pattern_with_group = r"(The) (\w+)"
match_with_group = re.match(pattern_with_group, text)
if match_with_group:
    print("match groups():", match_with_group.groups())
```

```python
# 6️⃣ sub() - replace 'Spain' with 'France'
print("sub():", re.sub(r"Spain", "France", text))
```

```python
# 7️⃣ compile() - compile pattern for reuse
pattern = re.compile(r"\b\w{5}\b")  # any 5-letter word
print("compile() + findall():", pattern.findall(text))
```

```python
text = """Hello World 123!
Regex demo: Cats, dogs, and birds.
Special symbols: @#$%_-
Line start here
And ends here
"""

print("=== Square brackets & Range ===")
print(re.findall(r"[aeiou]", text))         # all vowels
print(re.findall(r"[A-Z]", text))           # all uppercase letters
```

```python
print("\n=== Negation ===")
print(re.findall(r"[^A-Z]", "HelloWORLD"))  # not uppercase letters

print("\n=== OR (pipe) ===")
print(re.findall(r"cat|dog", "I love cats and dogs"))  # matches 'cat' or 'dog'
```

```python
print("\n=== Predefined classes ===")
print("Any char (.)     :", re.findall(r".", "Hi"))
print("Digits (\\d)     :", re.findall(r"\d", text))
print("Non-digits (\\D) :", re.findall(r"\D", "A1B2"))
print("Whitespace (\\s) :", re.findall(r"\s", "Hi there"))
print("Non-whitespace   :", re.findall(r"\S", "Hi there"))
print("Word chars (\\w) :", re.findall(r"\w", "Hi_123!"))
print("Non-word chars   :", re.findall(r"\W", "Hi_123!"))
```

```python
print("\n=== Boundary characters ===")
print("Start of line (^)", re.findall(r"^Hello", text, re.MULTILINE))
print("End of line ($)  :", re.findall(r"here$", text, re.MULTILINE))
print("Word boundary (\\b):", re.findall(r"\bcat\b", "cat scatter caterpillar"))
print("Non-word boundary:", re.findall(r"\Bcat\B", "concatenate"))
print("Start of input (\\A):", re.findall(r"\AHello", text))
print("End of prev match (\\G):", re.findall(r".", "abc"))  # Rarely used
print("End of input (\\Z):", re.findall(r"here\Z", text))
```

```python
#Standard text preprocessing

import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download NLTK data files (only need once)

#nltk.download('punkt')
#nltk.download('punkt_tab')
#nltk.download('stopwords')
#nltk.download('wordnet')
#nltk.download('omw-1.4')


# Sample text
text = "Cats are running faster than dogs, and John’s happiness was remarkable!"

# Step 1: Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)

# Step 2: Lowercasing
tokens = [word.lower() for word in tokens]
print("Lowercased:", tokens)

# Step 3: Stopword removal
stop_words = set(stopwords.words("english"))
tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
print("Stopword Removal:", tokens)

# Step 4: Stemming
stemmer = PorterStemmer()
stems = [stemmer.stem(word) for word in tokens]
print("Stemming:", stems)
```

```python
# Lemmatization using spaCy
import spacy

# Load English language model with NER capability
nlp = spacy.load("en_core_web_sm")

doc = nlp(" ".join(tokens))
lemmas = [token.lemma_ for token in doc]
print("Lemmatization:", lemmas)
```

```python
# NER
# Example text
text = "Apple is looking at buying a U.K. startup for $1 billion in 2025."
# Process the text
doc = nlp(text)

# Extract and print named entities
print(f"{'Entity':<20} {'Label':<15} {'Explanation'}")
print("-" * 60)
for ent in doc.ents:
    print(f"{ent.text:<20} {ent.label_:<15} {spacy.explain(ent.label_)}")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon (only once)
#nltk.download('vader_lexicon')

# Initialize the sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Sample sentences
sentences = [
    "I absolutely love this product! It's amazing 😍",
    "This is the worst experience I’ve ever had.",
    "It’s okay, not great but not bad either.",
    "The customer support is killing me!",
    "You are killing it! Great job!"
]
```

```python
# Analyze each sentence
for sentence in sentences:
    scores = sid.polarity_scores(sentence)
    print(f"\nSentence: {sentence}")
    print("Scores:", scores)

    # Simple classification
    if scores['compound'] >= 0.05:
        print("Sentiment: Positive")
    elif scores['compound'] <= -0.05:
        print("Sentiment: Negative")
    else:
        print("Sentiment: Neutral")
```

```python
# Text Classification

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# Sample dataset (positive & negative reviews)
texts = [
    "I love this movie, it’s fantastic and fun!",      # Positive
    "Absolutely terrible, I hated every moment.",      # Negative
    "What a great performance, really enjoyed it.",    # Positive
    "Worst film ever, boring and too long.",           # Negative
    "This was a wonderful experience, very touching.", # Positive
    "Disappointing story and bad acting.",             # Negative
]

labels = ["pos", "neg", "pos", "neg", "pos", "neg"]  # Corresponding labels
```

```python
# Build pipeline: Vectorizer + Classifier
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(texts, labels)
```

```python
# Test predictions
test_texts = [
    "The movie was fantastic, I loved it!",
    "It was boring and predictable.",
    "Such a heartwarming story!",
    "Terrible acting, waste of time."
]

predictions = model.predict(test_texts)

# Show results
for text, label in zip(test_texts, predictions):
    print(f"Text: {text}\nPredicted label: {label}\n")
```
